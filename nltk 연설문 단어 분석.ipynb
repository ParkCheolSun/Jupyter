```python
from nltk.book import *
from nltk import word_tokenize
from nltk import pos_tag
```

```python
file = open('Chapter2.txt','r')
text = file.read()
text = word_tokenize(text)
words = pos_tag(text)
nonus = ['NN','NNP','NNS','NNPS']
nonu_word = []
for word in words:
    for nonu in nonus:
        if word[1] == nonu:
            nonu_word.append(word[0])
#print(nonu_word)

word_count = {}
for word in nonu_word:
    if word not in word_count:
        word_count[word] = 1
    else:
        word_count[word] += 1
print('3번 이상 나온 명사')
for word, value in word_count.items():
    if value >= 3:
        print(word,':',value)
        
print()
print('3번 이상 나온 형용사와 동사')
ad_vb = ['JJ','JJR','JJS','VB','VBD','VBG','VBN','VBP','VBZ']
advb_word = []
for word in words:
    for advb in ad_vb:
        if word[1] == advb:
            advb_word.append(word[0])
advb_count = {}
for word in advb_word:
    if word not in advb_count:
        advb_count[word] = 1
    else:
        advb_count[word] += 1
for word, value in advb_count.items():
    if value >= 3:
        print(word,':',value)
```
