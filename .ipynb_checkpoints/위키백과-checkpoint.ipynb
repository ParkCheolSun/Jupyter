```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
def getLinks(url):
    #url = 'https://en.wikipedia.org/wiki/Kevin_Bacon'
    url = 'https://en.wikipedia.org{}'.format(url)
    html = urlopen(url)
    bs = BeautifulSoup(html, 'html.parser')
    links = bs.find(id='bodyContent').findAll('a')
    return links

def getContents(url):
    #url = 'https://en.wikipedia.org/wiki/Kevin_Bacon'
    url = 'https://en.wikipedia.org{}'.format(url)
    html = urlopen(url)
    bs = BeautifulSoup(html, 'html.parser')
    links = bs.find(id='bodyContent')
    return links
```

```python
link_List = []
for link in bs.find(id='bodyContent').findAll('a'):
    if 'href' in link.attrs:
        if link.attrs['href'][0] != '#':
            link_List.append(link.attrs['href'])
            
for link in link_List:
    print(link)
```
