```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
driver = webdriver.Chrome(executable_path='./chromedriver.exe')
url = 'https://www.naver.com'
driver.get(url)
time.sleep(3)
driver.find_element(By.XPATH,r'//*[@id="NM_FAVORITE"]/div[1]/ul[2]/li[3]/a').click()
time.sleep(3)
driver.find_element(By.XPATH,r'//*[@id="content"]/div[1]/div[1]/div[2]/div[1]/a/em').click()
time.sleep(3)
driver.execute_script("window.scrollTo(0, 3000)")
time.sleep(1)
main_lists1 = []
main_lists2 = []
# (3~7)(11~15)(19~23) 3 -> 11 -> 19
#방법 1
for i in range(3,160,8):
    for j in range(i,i+5):
        name = driver.find_element(By.XPATH,r'//*[@id="contentarea"]/div[3]/table/tbody/tr['+str(j)+']/td[2]/a').text
        cost = driver.find_element(By.XPATH,r'//*[@id="contentarea"]/div[3]/table/tbody/tr['+str(j)+']/td[3]').text
        sub_list = []
        sub_list.append(name)
        sub_list.append(cost)
        main_lists1.append(sub_list)
print(len(main_lists1))
for item in main_lists1:
    print('방법1',item)

#방법 2
for i in range(3,160):
    temp = i % 8
    if temp < 3:
        continue
    name = driver.find_element(By.XPATH,r'//*[@id="contentarea"]/div[3]/table/tbody/tr['+str(i)+']/td[2]/a').text
    cost = driver.find_element(By.XPATH,r'//*[@id="contentarea"]/div[3]/table/tbody/tr['+str(i)+']/td[3]').text
    sub_list = []
    sub_list.append(name)
    sub_list.append(cost)
    main_lists2.append(sub_list)
for item in main_lists2:
    print('방법2',item)
```
