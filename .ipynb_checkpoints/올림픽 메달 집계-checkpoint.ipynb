```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import unicodedata
import csv
```

```python
url = 'https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84'
html = urlopen(url)
bs = BeautifulSoup(html, 'html.parser')
```

```python
def fill_str_with_space(input_s="", max_size=40, fill_char=" "):
    """
    - 길이가 긴 문자는 2칸으로 체크하고, 짧으면 1칸으로 체크함. 
    - 최대 길이(max_size)는 40이며, input_s의 실제 길이가 이보다 짧으면 
    남은 문자를 fill_char로 채운다.
    """
    l = 0 
    for c in input_s:
        if unicodedata.east_asian_width(c) in ['F', 'W']: # 'F'=Fullwitdh / 'W'=Wide / 'H'=Halfwidth
            l+=2
        else: 
            l+=1
    return input_s+fill_char*(max_size-l)
```

```python
f = open('medal.csv','w',newline='')
wr = csv.writer(f)
datas = bs.select('#mw-content-text > div.mw-parser-output > table:nth-child(4) > tbody > tr')
#print(datas)
cnt = 0
for data in datas:
    cnt += 1
    if cnt == 1:
        continue
    try:
        loc = data.select_one('td a').string
        gold = data.select_one('td:nth-child(3)').string
        silver = data.select_one('td:nth-child(4)').string
        bronze = data.select_one('td:nth-child(5)').string
        print('---------------------------------------------------------------')
        print('{:15s} {:^5s} {:^5s} {:^5s}'.format(fill_str_with_space(loc, max_size=40),gold,silver,bronze))
        wr.writerow([loc,gold,silver,bronze])
    except:
        print('---------------------------------------------------------------')
        print('더이상 데이터가 없습니다.')

f.close()
```
