```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import re
driver = webdriver.Chrome(executable_path='./chromedriver.exe')
url = 'https://petitions.assembly.go.kr/'
driver.get(url)
time.sleep(3)
driver.find_element(By.XPATH,r'//*[@id="mainMenu_list"]/li[3]/div[1]/a').click()
time.sleep(3)
driver.find_element(By.XPATH,r'//*[@id="navsubbar"]/ul/li[2]/a').click()
time.sleep(3)
text = ''
for i in range(1,9):
    driver.find_element(By.XPATH,r'//*[@id="contentsbody"]/div/div/div[4]/div[2]/div[2]/ul/li['+str(i)+']/div/a/dl/dd').click()
    time.sleep(2)
    text += driver.find_element(By.XPATH,r'//*[@id="contentsbody"]/div/div/div[3]/div[2]/dl[3]/dd').text
    driver.back()
    
print(text)
```
