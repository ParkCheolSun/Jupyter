```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
html = urlopen('https://www.unicef.org/press-releases/we-have-learned-love-ourselves-so-now-i-urge-you-speak-yourself')
bs = BeautifulSoup(html, 'html.parser')
```

```python
def word_list(list):
    words = []
    word_list = []
    for li in list:
        li = re.sub(r'\s+',' ',li)
        li = li.replace('\n','')
        li = li.replace('\r','')
        li = li.replace('\"','')
        li = li.replace('#','')
        li = li.replace('!','')
        li = li.replace(',','')
        li = li.replace('.','')
        li = li.replace('‘','')
        li = li.replace('’','')
        li = li.replace('“','')
        li = li.replace('”','')
        words = li.split(' ')
        word_list.append(words)
        
    flat_list = []
    for l in word_list:
        for item in l:
            flat_list.append(item)
            
    word_dict = {}
    for word in flat_list:
        if word not in word_dict:
            word_dict[word] = 1
        else:
            word_dict[word] += 1
    for key, values in word_dict.items():
        print(key,values)
```

```python
title = bs.find(class_='h1')
print(title.string)

sub_scripts = bs.select('#block-unicef-base-mainpagecontent > article > div.main-content-cntr.selectric-small > div > div > div.press-release--left-rail > div.press_text_cntr > div > div > div > div > div')
sub_list = []
for sub in sub_scripts:
    sub_list.append(sub.get_text())
    
#print(sub_list)
word_list(sub_list)
```
