```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
from urllib.request import HTTPError
from urllib.request import URLError
url = 'http://www.pythonscraping.com/pages/page1.html'
try:
    html = urlopen(url)
except HTTPError as e:
    print('웹에 접속할 수 없습니다.')
    print(e)
except URLError as e:
    print('웹 주소를 찾을 수 없습니다.')
    print(e)
else :
    bs = BeautifulSoup(html.read(), 'html.parser')
    print('Title : ' + bs.h1.string)
    print('Contents : ')
    print(bs.div.string)
```

```python
url = 'https://www.pythonscraping.com/pages/warandpeace.html'
html = urlopen(url)
bs = BeautifulSoup(html.read(),'html.parser')
nameList = bs.findAll('span',{'class' : {'green','red'}})
#print(nameList)
for name in nameList:
    print(name.string)
```

```python
url = 'https://www.naver.com'
html = urlopen(url)
bs = BeautifulSoup(html.read(),'html.parser')
#startPage = bs.find(id='NM_set_home_btn')
#print(startPage.string)
#titles = bs.findAll(class_='title')
titles = bs.findAll('strong',{'class':'title'})
for title in titles:
    print(title.string)
```

```python
url = 'https://finance.naver.com'
html = urlopen(url)
bs = BeautifulSoup(html.read(),'html.parser')
titles1 = bs.find(class_='section_strategy')
titles2 = titles1.find('ul')
titles3 = titles2.findAll('a')
for title in titles3:
    print(title.string)
    print()
```

```python
url = 'https://pythonscraping.com/pages/page3.html'
html = urlopen(url)
bs = BeautifulSoup(html.read(),'html.parser')
print(bs)
```

```python
childs = bs.find('table',{'id':'giftList'}).children
for ch in childs:
    print(ch)
```

```python
sibling = bs.find('table',{'id':'giftList'}).tr.next_siblings
for sib in sibling:
    print(sib)
```

```python
parent_text = bs.find('img',{'src':'../img/gifts/img1.jpg'}).parent.previous_sibling
print(parent_text.string)
```

```python
import re
images = bs.findAll('img',{'src':re.compile('\.\.\/img\/gifts/img.*\.jpg')})
for image in images:
    print(image)
    print(image['src'])
```
