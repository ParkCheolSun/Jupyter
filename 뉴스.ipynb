```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
url_mbc = 'https://imnews.imbc.com/replay/2022/nwdesk/'
url_kbs = 'https://news.kbs.co.kr/news/list.do?mcd=0795&ref=pSiteMap#1'
url_sbs = 'https://news.sbs.co.kr/news/newsMain.do'
```

```python
def getUrl(url):
    html = urlopen(url)
    bs = BeautifulSoup(html, 'html.parser')
    return bs
```

```python
def getMBCContents(bs):
    datas = bs.select('#content > section > div.section.content_area > div.list_area.list_top.wrapper')
    for data in datas:
        tit = data.findAll(class_='tit')
        sub = data.findAll(class_='sub')
        for cnt in range(len(tit)):
            print('제목 :','['+tit[cnt].string+']')
            print(sub[cnt].string)
            print()
            
def getMBC(bs):
    bs = getUrl('https://imnews.imbc.com/news/2022/politics/article/6407914_35666.html')
    datas = bs.find(class_='news_txt')
    data_list = []
    for i in datas:
        temp = re.sub('<.+?>', '', str(i), 0).strip()
        if temp == '':
            continue
        data_list.append(temp)
    return data_list
    
            
def getKBSContents(bs):
    bs = getUrl(url_kbs)
    tit_datas = bs.findAll(class_='tit')
    sub_datas = bs.findAll(class_='script')
    try:
        for cnt in range(len(tit_datas)):
            print('제목 :',tit_datas[cnt].string)
            print(sub_datas[cnt].string)
    except:
        print('[스크립트로 작성되어있습니다.]')

def getSBSContents(bs):
    url = 'https://news.sbs.co.kr/news{}'.format(url)
    html = urlopen(url)
    bs = BeautifulSoup(html, 'html.parser')
    data = bs.find(id='bodyContent')
    return data

def word_list(list):
    words = []
    word_list = []
    for li in list:
        li = re.sub(r'\s+',' ',li)
        li = li.replace('\n','')
        li = li.replace('\r','')
        li = li.replace('\"','')
        words = li.split(' ')
        word_list.append(words)
        
    flat_list = []
    for l in word_list:
        for item in l:
            flat_list.append(item)
            
    word_dict = {}
    for word in flat_list:
        if word not in word_dict:
            word_dict[word] = 1
        else:
            word_dict[word] += 1
    for key, values in word_dict.items():
        print(key,values)
```

```python
number = 2
if number == 0:
    getMBCContents(getUrl(url_mbc))
elif number == 1:
    getKBSContents(getUrl(url_kbs))
    
list = getMBC(getUrl(url_mbc))
print(list)
word_list(list)
```
